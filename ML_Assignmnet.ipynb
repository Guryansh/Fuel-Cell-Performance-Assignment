```python
from pycaret.regression import *
import pandas as pd
```

```python
data = pd.read_csv('Fuel_cell_performance_data-Full.csv')
```

```python
selected_data = data[[data.columns[i] for i in range(15)]+['Target5']]
selected_data.head()
model_df = pd.DataFrame(columns=['Model Code', 'Model Parameters', 'R2'])
```

# **1.1 Model Using 10 Folds**

```python
setup(data = selected_data, target='Target5', train_size = 0.7, data_split_shuffle = False, verbose=False)
cm = compare_models()
comparison_results = pull()
r2_score = comparison_results.loc[comparison_results.index[0], 'R2']
top_model = cm
model_df.loc[len(model_df)] = [top_model.__class__.__name__, 'None, 10 folds', r2_score]
```
